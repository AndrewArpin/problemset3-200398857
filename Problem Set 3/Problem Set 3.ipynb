```python
#step 1
#step 2
#step 3
import numpy as np
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep = '|')
#step 4
data.groupby('occupation')['age'].mean()
#step 5
data['male'] = data['gender'].apply(lambda x: 1 if x == 'M' else 0)
totals = data.groupby('occupation')['male'].sum() 
(totals / data.groupby('occupation')['gender'].count())
#step 6
data.groupby('occupation')['age'].max()
data.groupby('occupation')['age'].min()
#step 7
data.groupby(['occupation', 'gender'])['age'].mean()
#step 8
data['male'] = data['gender'].apply(lambda x: 1 if x == 'M' else 0)
data['female'] = data['gender'].apply(lambda x: 1 if x == 'F' else 0)

totalM = data.groupby('occupation')['male'].sum() 
totalF = data.groupby('occupation')['female'].sum() 
M = (totalM / data.groupby('occupation')['gender'].count())
F = (totalF / data.groupby('occupation')['gender'].count())
print('total males', M)
print('total females', F)
```

```python
# Step 1. Import the necessary libraries
# Step 2. Import the dataset from this address
# Step 3. Assign it to a variable called euro12
import numpy as np
import pandas as pd
euro12 = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv')

#Step 4. Select only the Goal column
goals = euro12["Goals"]
print(goals)

#Step 5. How many team participated in the Euro2012?
Teams = len(euro12.index)
print("There are " + str(Teams) + " Teams")

#Step 6. What is the number of columns in the dataset?
Columns = len(euro12.columns)
print("There are " + str(Columns) +" columns")

#Step 7. View only the columns Team, Yellow Cards and Red Cards and assign them to a dataframe called discipline
#Step 8. Sort the teams by Red Cards, then to Yellow Cards
discipline = euro12.sort_values(["Red Cards","Yellow Cards"], ascending = (False,False))
print(discipline)

#Step 9. Calculate the mean Yellow Cards given per Team
yellow = euro12[["Yellow Cards"]]
print(yellow.mean())

#Step 10. Filter teams that scored more than 6 goalsStep 11. Select the teams that start with G
abc = euro12[(euro12["Goals"]>6) & (euro12["Team"].str.startswith("G"))]
print(abc)
           
#Step 12. Select the first 7 columns
firstcolumns = euro12.iloc[:, :7]
print(firstcolumns)

#Step 13. Select all columns except the last 3
lastcolumn = euro12.iloc[:, -3:]
print(lastcolumn) 

# #Step 14. Present only the Shooting Accuracy from England, Italy and Russia
shootingacc = euro12[(euro12["Team"]=="England") | (euro12["Team"]=="Italy")| (euro12["Team"]=="Russia")] 
shootingacc = shootingacc["Shooting Accuracy"]
print(shootingacc)
```

```python
import matplotlib.pyplot as plt
data = pd.read_csv("roman-emperor-reigns.csv")
l = data['Cause_of_Death'].unique()
print(l)
a = data['Cause_of_Death'].apply(lambda x: 1 if x == 'Possibly assassinated' else 0)
b = data['Cause_of_Death'].apply(lambda x: 1 if x == 'Illness' else 0)
c = data['Cause_of_Death'].apply(lambda x: 1 if x == 'Assassinated' else 0)
d = data['Cause_of_Death'].apply(lambda x: 1 if x == 'Killed in battle' else 0)
e = data['Cause_of_Death'].apply(lambda x: 1 if x == 'Executed' else 0)
f = data['Cause_of_Death'].apply(lambda x: 1 if x == 'Natural causes' else 0)
g = data['Cause_of_Death'].apply(lambda x: 1 if x == 'Died in captivity' else 0)
h = data['Cause_of_Death'].apply(lambda x: 1 if x == 'Suicide' else 0)

data =[a,b,c,d,e,f,g,h]
lables2 = 'Possibly assassinated','Illness','Assassinated','Killed in battle','Executed','Natural causes','Died in captivity','Suicide'
plt.pie(data, labels=lables2,autopct='%1.1f%%')
plt.title('My Title')
plt.axis('equal')
plt.show()
```
